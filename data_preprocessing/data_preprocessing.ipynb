```python
# Aliases for clarity
df_customer         = maps["df_customer"]
df_geolocation      = maps["df_geolocation"]
df_order_item       = maps["df_order_item"]
df_order_payment    = maps["df_order_payment"]
df_order_review     = maps["df_order_review"]
df_orders           = maps["df_orders"]
df_product          = maps["df_product"]
df_sellers          = maps["df_sellers"]
df_product_name     = maps["df_product_name"]  # Contains product_category_name translations

# Step 1: Merge orders with order items
df_merged = pd.merge(df_orders, df_order_item, on="order_id", how="inner")

# Step 2: Merge with products
df_merged = pd.merge(df_merged, df_product, on="product_id", how="left")

# Step 3: Enrich with translated product category name
df_merged = pd.merge(
    df_merged,
    df_product_name,
    on="product_category_name",
    how="left"
)

# Step 4: Merge with customer data
df_merged = pd.merge(df_merged, df_customer, on="customer_id", how="left")

# Step 5: Merge with seller data
df_merged = pd.merge(df_merged, df_sellers, on="seller_id", how="left")

# Step 6: Merge with review scores (only retain score)
df_merged = pd.merge(
    df_merged,
    df_order_review[["order_id", "review_score"]],
    on="order_id",
    how="left"
)

# Step 7: Merge with payment data
df_merged = pd.merge(df_merged, df_order_payment, on="order_id", how="left")

# Final output
df_merged.to_csv("../data_processed/olist_combined_clean_with_product_name.csv", index=False)
print("âœ… Final dataset with translated product names saved. Shape:", df_merged.shape)
```

```python
# Columns safe to drop due to nulls or low business value
cols_to_drop = [
    "product_category_name_english", 
    "payment_sequential", 
    "payment_installments",
    
]

# Optional: Only drop payment_type if confirmed non-critical
if df_merged["payment_type"].isnull().sum() > 0:
    cols_to_drop.append("payment_type")

# Apply column drop
df_merged.drop(columns=cols_to_drop, inplace=True)

print("âœ… Dropped columns:", cols_to_drop)

```

```python
for col in ["product_weight_g", "product_length_cm", "product_height_cm", "product_width_cm"]:
    df_merged[col] = df_merged[col].fillna(df_merged[col].median())

# For review_score (e.g., fill missing with -1)
df_merged["review_score"] = df_merged["review_score"].fillna(-1)

# For payment_value (fill with median)
df_merged["payment_value"] = df_merged["payment_value"].fillna(df_merged["payment_value"].median())
```

```python
# 1. Identify duplicate order_ids (excluding the first occurrence)
duplicate_mask = df_merged["order_id"].duplicated(keep=False)

# 2. Filter rows with duplicate order_ids
duplicates_df = df_merged[duplicate_mask]

# 3. Summarize the associated financial columns
summary = duplicates_df[["payment_value", "freight_value", "price"]].describe()

# 4. Count of duplicate order_ids
num_duplicates = df_merged["order_id"].duplicated().sum()

# 5. Optional: Group by order_id to inspect variations
grouped_summary = duplicates_df.groupby("order_id")[["payment_value", "freight_value", "price"]].agg(["count", "mean", "sum", "std"])

# Output
print(f"ðŸ” Number of duplicate order_id entries: {num_duplicates}\n")
print("ðŸ“Š Summary of financials for duplicate order_ids:")
print(summary)

# Optional detailed view
print(grouped_summary)

display(duplicates_df[["order_id","product_id", "payment_value", "freight_value", "price"]])
```
